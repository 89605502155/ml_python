```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Генерируем случайные данные
np.random.seed(0)
x = np.linspace(0, 10, 100)
y_true = 2 * x + 1
noise = np.random.normal(0, 1, size=len(x))
y = y_true + noise

# Оцениваем параметры линейной регрессии
slope, intercept = np.polyfit(x, y, 1)

# Рисуем данные и прямую
plt.scatter(x, y, label='Данные')
plt.plot(x, slope * x + intercept, color='red', label='Прямая')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Прямая и данные')
plt.legend()

# Вычисляем доверительный интервал для параметра наклона (slope)
# (предполагаем, что параметры линейной регрессии нормально распределены)
std_error = np.sqrt(np.sum((y - (slope * x + intercept))**2) / (len(x) - 2))
z_score = norm.ppf(0.975)  # 95% доверительный интервал
margin_of_error = z_score * std_error
confidence_interval = (slope - margin_of_error, slope + margin_of_error)

# Рисуем доверительный интервал
plt.fill_between(x, (slope - margin_of_error) * x + intercept, (slope + margin_of_error) * x + intercept,
                 color='gray', alpha=0.2, label='Доверительный интервал')

plt.show()

print("Доверительный интервал для параметра наклона:", confidence_interval)
```
