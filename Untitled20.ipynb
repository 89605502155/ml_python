```python
import numpy as np
import math
```

```python
x = np.arange(0,4.01,0.01)
```

```python
class qv:
  def __init__(self,a) -> None:
    self.al=a

  def kor(self):
    return math.sqrt(math.pow(self.b,2)-(4*self.a*self.c))

  def main(self):
    self.a=1
    self.b=self.al-6
    self.c=math.pow(self.al-1,2)-(4*(self.al-1))+4
    x1=(((-1)*self.b)+self.kor())/(2*self.a)
    x2=(((-1)*self.b)-self.kor())/(2*self.a)
    if x2+self.al<3:
      return x1,0
    else:
      return x1,x2
```

```python
class ger_arr:
  def __init__(self,x:np.ndarray) -> None:
    self.x=x

  def main(self):
    y1=np.zeros(self.x.shape[0])
    y2=np.zeros(self.x.shape[0])
    for i in range(self.x.shape[0]):
      mod=qv(self.x[i])
      y1[i],y2[i]=mod.main()
    return y1,y2
```

```python
mod=ger_arr(x)
y1,y2=mod.main()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rc('font',family='Times New Roman')
fig = plt.figure(figsize=(9, 9.5),constrained_layout=True)
ax=fig.add_subplot()
plt.plot(x, y1,'green')
plt.plot(x, y2,'green')
plt.fill_between(x, y1,y2,color='green')
ax.set_ylabel("$\\alpha$", fontsize=40)
ax.set_xlabel("$\\beta$", fontsize=40)
#plt.tight_layout(h_pad=1,w_pad=5)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=30)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=30)
plt.grid()
ax.get_xaxis().set_tick_params(direction='in')
ax.get_yaxis().set_tick_params(direction='in')
save_file_name="cubic_interp"
plt.savefig(save_file_name+'.pdf', format='pdf', dpi=300,bbox_inches='tight')
plt.savefig(save_file_name+'.jpg', format='jpg', dpi=300,bbox_inches='tight')
plt.savefig(save_file_name+".svg", format="svg",bbox_inches='tight')
plt.show()
```
