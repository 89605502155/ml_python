```python
class qv:
  def __init__(self,a) -> None:
    self.al=a

  def kor(self):
    return math.sqrt(math.pow(self.b,2)-(4*self.a*self.c))

  def main(self):
    self.a=1
    self.b=self.al-6
    self.c=math.pow(self.al-1,2)-(4*(self.al-1))+4
    x1=(((-1)*self.b)+self.kor())/(2*self.a)
    x2=(((-1)*self.b)-self.kor())/(2*self.a)
    if x2+self.al<3:
      return x1,0
    else:
      return x1,x2
```

```python
class ger_arr:
  def __init__(self,x:np.ndarray) -> None:
    self.x=x

  def main(self):
    y1=np.zeros(self.x.shape[0])
    y2=np.zeros(self.x.shape[0])
    for i in range(self.x.shape[0]):
      mod=qv(self.x[i])
      y1[i],y2[i]=mod.main()
    return y1,y2
```

```python
class ParzenWindowClassifier:
    def __init__(self, h=0.5):
        self.h = h  # Ширина окна
    
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        self.classes = np.unique(y_train)
        
        
    
    def predict(self, X_test):
        predictions = []
        for x in X_test:
            # Вычисляем расстояние от тестового объекта до всех точек обучающей выборки
            distances = np.linalg.norm(self.X_train - x, axis=1)
            # Вычисляем ядерную оценку для каждого расстояния
            kernel_values = np.exp(-0.5 * (distances / self.h) ** 2) / (np.sqrt(2 * np.pi) * self.h)
            # Вычисляем взвешенную сумму ядерных оценок для каждого класса
            class_scores = {c: np.sum(kernel_values[self.y_train == c]) for c in self.classes}
            # Предсказываем класс с наибольшей взвешенной суммой ядерных оценок
            prediction = max(class_scores, key=class_scores.get)
            predictions.append(prediction)
        return np.array(predictions)
```

```python
import numpy as np
from typing import Literal
from sklearn.base import BaseEstimator
from sklearn.base import RegressorMixin

class parzen_window_classifier(BaseEstimator,RegressorMixin):
    def __init__(self, h:float=0.5,kernel:Literal['gaus', 'epanichenkov', 'quadr',
                                            'triangle','rectangle']  = "gaus"):
        self.h = h 
        self.kernel=kernel

    def __epanichenkov(self,r):
        res=[]
        for i in r:
            if abs(i)<=1:
                res.append((0.75)*(1-(i**2)))
            else:
                res.append(0)
        return np.array(res)
    
    def __quadr(self,r):
        res=[]
        for i in r:
            if abs(i)<=1:
                res.append((15.0/16.0)*((1-(i**2))**2))
            else:
                res.append(0)
        return np.array(res)
    
    def __triangle(self,r):
        res=[]
        for i in r:
            if abs(i)<=1:
                res.append(1-abs(i))
            else:
                res.append(0)
        return np.array(res)
    
    def __rectangle(self,r):
        res=[]
        for i in r:
            if abs(i)<=1:
                res.append(0.5)
            else:
                res.append(0)
        return np.array(res)
    
    def __set_kernel(self,distance:float):
        match self.kernel:
            case 'gaus': return np.exp(-0.5 * (distance / self.h) ** 2) / (np.sqrt(2 * np.pi) * self.h)
            case 'epanichenkov': return  self.__epanichenkov(distance / self.h)
            case 'quadr': return self.__quadr(distance / self.h)
            case 'triangle': return self.__triangle(distance / self.h)
            case 'rectangle': return self.__rectangle(distance / self.h)

        
    
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        self.classes = np.unique(y_train)
        
    
    def predict(self, X_test):
        predictions = []
        for x in X_test:
            # Вычисляем расстояние от тестового объекта до всех точек обучающей выборки
            distances = np.linalg.norm(self.X_train - x, axis=1)
            # Вычисляем ядерную оценку для каждого расстояния
            # kernel_values = np.array([self.__set_kernel(distance=i) for i in distances])
            kernel_values = self.__set_kernel(distance=distances)
            # Вычисляем взвешенную сумму ядерных оценок для каждого класса
            class_scores = {c: np.sum(kernel_values[self.y_train == c]) for c in self.classes}
            # Предсказываем класс с наибольшей взвешенной суммой ядерных оценок
            prediction = max(class_scores, key=class_scores.get)
            predictions.append(prediction)
        return np.array(predictions)
```

```python
def plot_decision_boundary(classifier, X, y):
    h = 2  # Шаг сетки
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[y == 0][:, 0], X[y == 0][:, 0], label='Sm', edgecolors='k', cmap=plt.cm.Paired)
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 0], label='Noise', edgecolors='k', cmap=plt.cm.Paired)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Parzen Window Classifier Decision Boundary')
```

```python
import gzip
import pickle
import zipfile
import os

def convert_pkl_gz_to_zip(pkl_gz_file, zip_file):
    # Чтение и распаковка файла .pkl.gz
    with gzip.open(pkl_gz_file, 'rb') as f:
        data = pickle.load(f)

    # Создание файла .zip
    with zipfile.ZipFile(zip_file, 'w') as zf:
        # Добавление распакованных данных в архив
        zf.writestr(os.path.basename(pkl_gz_file)[:-3], pickle.dumps(data))

# Укажите путь к вашему файлу .pkl.gz и путь для сохранения файла .zip
pkl_gz_file = 'smoother_of_loadings.pkl.gz'
zip_file = 'smoother_of_loadings.zip'

# Вызов функции для конвертации
convert_pkl_gz_to_zip(pkl_gz_file, zip_file)
```
